```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados.csv', sep=';')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
sns.histplot(df.PIB, kde=True, bins=10).set(title='PIB')
```

```python
sns.histplot(df.VALOREMPENHO, kde=True, bins=10).set(title='Valor de Empenho')
```

```python
agrupado = df.sort_values('PIB', ascending=False).head(10)
agrupado = agrupado.iloc[:,1:3]
ax = agrupado.plot.bar(x='MUNICIPIO', y='PIB', color = 'gray', legend = False)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=6)
plt.show()
```

```python
agrupado = df.sort_values('VALOREMPENHO', ascending=False).head(10)
agrupado = agrupado.iloc[:,[1,3]]
ax = agrupado.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color = 'gray', legend = False)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=6)
plt.show()
```

```python
#Criando atributo de proporção de valor empenho por PIB
df['PROPORCAO'] = round(df.VALOREMPENHO / df.PIB, 2)
df.head()
```

```python
agrupado = df.sort_values('PROPORCAO', ascending=False).head(10)
agrupado = agrupado.iloc[:,[1,4]]
ax = agrupado.plot.bar(x='MUNICIPIO', y='PROPORCAO', color = 'gray', legend = False)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=6)
plt.show()
```
